Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
def plot_pie_chart(column_name):
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Menggunakan palet warna pastel dari seaborn
    colors = sns.color_palette('pastel')[0:len(labels)]

    plt.figure(figsize=(8, 8))  # Membuat figure lebih besar untuk estetika
    wedges, texts, autotexts = plt.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        shadow=True,  # Menambahkan bayangan untuk membuatnya terlihat 3D
        explode=[0.05] * len(labels),  # Membuat irisan sedikit terpisah untuk menonjolkan estetika
        wedgeprops={'edgecolor': 'black'}  # Menambahkan edge untuk setiap irisan
    )

    # Mengatur estetika teks label
    for text in texts:
        text.set_fontsize(12)
        text.set_fontweight('bold')

    # Mengatur estetika teks persentase dalam irisan
    for autotext in autotexts:
        autotext.set_color('black')
        autotext.set_fontsize(12)
        autotext.set_fontweight('bold')

    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=16, fontweight='bold', color='navy')
    plt.tight_layout()
    plt.show()

# Menampilkan pie chart untuk beberapa kolom
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

````python
# Set up seaborn style for aesthetics
sns.set(style="whitegrid")

# 1. Membuat histogram untuk distribusi nilai math, reading, dan writing
plt.figure(figsize=(18, 6))  # Ukuran plot

plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True, color='blue', bins=20)
plt.title('Distribusi Math Score', fontsize=14)
plt.xlabel('Math Score', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)

plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True, color='green', bins=20)
plt.title('Distribusi Reading Score', fontsize=14)
plt.xlabel('Reading Score', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)

plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True, color='red', bins=20)
plt.title('Distribusi Writing Score', fontsize=14)
plt.xlabel('Writing Score', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)

plt.tight_layout()
plt.show()

# 2. Menggunakan KDE Plot untuk lebih detail distribusi nilai math, reading, dan writing
plt.figure(figsize=(12, 6))

sns.kdeplot(df['math score'], label='Math Score', color='blue', shade=True)
sns.kdeplot(df['reading score'], label='Reading Score', color='green', shade=True)
sns.kdeplot(df['writing score'], label='Writing Score', color='red', shade=True)

plt.title('Distribusi KDE Math, Reading, dan Writing Score', fontsize=16)
plt.xlabel('Score', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(True)

plt.show()
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
# Define the passing score
passing_score = 75

# 1. Menambahkan kolom baru untuk menandakan apakah siswa pass atau fail di setiap pelajaran
df['math_pass_status'] = df['math score'].apply(lambda x: 'Pass' if x >= passing_score else 'Fail')
df['reading_pass_status'] = df['reading score'].apply(lambda x: 'Pass' if x >= passing_score else 'Fail')
df['writing_pass_status'] = df['writing score'].apply(lambda x: 'Pass' if x >= passing_score else 'Fail')

# 2. Menghitung jumlah siswa yang pass dan fail untuk setiap pelajaran
math_pass_fail_counts = df['math_pass_status'].value_counts()
reading_pass_fail_counts = df['reading_pass_status'].value_counts()
writing_pass_fail_counts = df['writing_pass_status'].value_counts()

# 3. Menampilkan hasilnya
print("Math Pass/Fail Counts:")
print(math_pass_fail_counts)
print("\nReading Pass/Fail Counts:")
print(reading_pass_fail_counts)
print("\nWriting Pass/Fail Counts:")
print(writing_pass_fail_counts)

# 4. Visualisasi jumlah siswa yang Pass/Fail dalam bentuk bar chart
import matplotlib.pyplot as plt
import seaborn as sns

# Setup Seaborn style
sns.set(style="whitegrid")

# Membuat bar plot untuk setiap subject
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Math pass/fail bar plot
sns.countplot(x='math_pass_status', data=df, palette='Blues', ax=axes[0])
axes[0].set_title('Math Pass/Fail Counts')
axes[0].set_xlabel('Status')
axes[0].set_ylabel('Jumlah Siswa')

# Reading pass/fail bar plot
sns.countplot(x='reading_pass_status', data=df, palette='Greens', ax=axes[1])
axes[1].set_title('Reading Pass/Fail Counts')
axes[1].set_xlabel('Status')
axes[1].set_ylabel('Jumlah Siswa')

# Writing pass/fail bar plot
sns.countplot(x='writing_pass_status', data=df, palette='Reds', ax=axes[2])
axes[2].set_title('Writing Pass/Fail Counts')
axes[2].set_xlabel('Status')
axes[2].set_ylabel('Jumlah Siswa')

# Menampilkan plot
plt.tight_layout()
plt.show()
````

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
# Fungsi untuk membuat bar chart
def plot_bar_chart(x, y, hue, data, title):
    plt.figure(figsize=(12, 6))
    sns.barplot(x=x, y=y, hue=hue, data=data, ci=None)
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title=hue)
    plt.xticks(rotation=45)
    plt.show()

# 1. Nilai Matematika
plot_bar_chart('gender', 'math score', 'race/ethnicity', df, 'Average Math Score by Gender and Race/Ethnicity')

# 2. Nilai Membaca
plot_bar_chart('gender', 'reading score', 'race/ethnicity', df, 'Average Reading Score by Gender and Race/Ethnicity')

# 3. Nilai Menulis
plot_bar_chart('gender', 'writing score', 'race/ethnicity', df, 'Average Writing Score by Gender and Race/Ethnicity')

# 4. Pengaruh Tingkat Pendidikan Orang Tua
plot_bar_chart('parental level of education', 'math score', 'gender', df, 'Average Math Score by Parental Education Level and Gender')
plot_bar_chart('parental level of education', 'reading score', 'gender', df, 'Average Reading Score by Parental Education Level and Gender')
plot_bar_chart('parental level of education', 'writing score', 'gender', df, 'Average Writing Score by Parental Education Level and Gender')

# 5. Pengaruh Jenis Makan Siang
plot_bar_chart('lunch', 'math score', 'gender', df, 'Average Math Score by Lunch Type and Gender')
plot_bar_chart('lunch', 'reading score', 'gender', df, 'Average Reading Score by Lunch Type and Gender')
plot_bar_chart('lunch', 'writing score', 'gender', df, 'Average Writing Score by Lunch Type and Gender')

# 6. Pengaruh Kursus Persiapan Ujian
plot_bar_chart('test preparation course', 'math score', 'gender', df, 'Average Math Score by Test Preparation Course and Gender')
plot_bar_chart('test preparation course', 'reading score', 'gender', df, 'Average Reading Score by Test Preparation Course and Gender')
plot_bar_chart('test preparation course', 'writing score', 'gender', df, 'Average Writing Score by Test Preparation Course and Gender')
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
# Menghitung nilai rata-rata dari seluruh mata pelajaran
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Mengatur tampilan seaborn
sns.set(style="whitegrid")

# Fungsi untuk membuat boxplot
def plot_boxplot(x, y, data, title):
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=x, y=y, data=data)
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.xticks(rotation=45)
    plt.show()

# Fungsi untuk membuat violinplot
def plot_violinplot(x, y, data, title):
    plt.figure(figsize=(12, 6))
    sns.violinplot(x=x, y=y, data=data)
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.xticks(rotation=45)
    plt.show()

# 1. Boxplot untuk nilai rata-rata berdasarkan gender
plot_boxplot('gender', 'average score', df, 'Average Score by Gender')

# 2. Violinplot untuk nilai rata-rata berdasarkan gender
plot_violinplot('gender', 'average score', df, 'Average Score Distribution by Gender')

# 3. Boxplot untuk nilai rata-rata berdasarkan ras/etnis
plot_boxplot('race/ethnicity', 'average score', df, 'Average Score by Race/Ethnicity')

# 4. Violinplot untuk nilai rata-rata berdasarkan ras/etnis
plot_violinplot('race/ethnicity', 'average score', df, 'Average Score Distribution by Race/Ethnicity')

# 5. Boxplot untuk nilai rata-rata berdasarkan tingkat pendidikan orang tua
plot_boxplot('parental level of education', 'average score', df, 'Average Score by Parental Education Level')

# 6. Violinplot untuk nilai rata-rata berdasarkan tingkat pendidikan orang tua
plot_violinplot('parental level of education', 'average score', df, 'Average Score Distribution by Parental Education Level')

# 7. Boxplot untuk nilai rata-rata berdasarkan jenis makan siang
plot_boxplot('lunch', 'average score', df, 'Average Score by Lunch Type')

# 8. Violinplot untuk nilai rata-rata berdasarkan jenis makan siang
plot_violinplot('lunch', 'average score', df, 'Average Score Distribution by Lunch Type')

# 9. Boxplot untuk nilai rata-rata berdasarkan kursus persiapan ujian
plot_boxplot('test preparation course', 'average score', df, 'Average Score by Test Preparation Course')

# 10. Violinplot untuk nilai rata-rata berdasarkan kursus persiapan ujian
plot_violinplot('test preparation course', 'average score', df, 'Average Score Distribution by Test Preparation Course')
````
